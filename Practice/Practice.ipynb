```python
import numpy as np
import matplotlib.pyplot as plt

def unit_step(n):
    return np.where(n >= 0, 1, 0)

def unit_ramp(n):
    return np.where(n >= 0, n, 0)

def exponential(n, a = 0.5):
    return np.exp(a * n)

def sine(n, f = 1):
    return np.sin(2 * np.pi * f * n / 10)

def cosine(n, f = 1):
    return np.cos(2 * np.pi * f * n /10)

n = np.arange(-10, 11, 1)

signals = {
    "Unit Step" : unit_step(n),
    "Unit Ramp" : unit_ramp(n),
    "Exponential" : exponential(n),
    "Sine" : sine(n),
    "Cosine" : cosine(n)
}

plt.figure(figsize=(12, 8))
for name, function in signals.items():
    plt.subplot(3, 2, list(signals.keys()).index(name) + 1)
    plt.stem(n, function, label=name)
    plt.title(name)
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.grid()
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def signals(n , f = 10):
    return np.sin(2 * np.pi * n * f)

n = np.linspace(0, 1, 1000)
n_sample = np.arange(0 , 1 , 1/ 50)
n_alisging = np.arange(0, 1, 1 / 15)

plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(n, signals(n), label='Sine Wave')
plt.stem(n_sample, signals(n_sample), linefmt='red', label='Sampled Points')
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()

plt.subplot(2, 1, 2)
plt.plot(n, signals(n), label='Sine Wave')
plt.stem(n_alisging, signals(n_alisging), linefmt='green', label='Aliased Points')
plt.title('Aliased Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def unit_stem(n):
    return np.where(n >= 0, 1, 0)

n = np.arange(0 , 10, 1)

h = unit_stem(n) - unit_stem(n - 5)
x = unit_stem(n)

y = np.convolve(x, h, mode='full')


plt.figure(figsize=(10, 5))
plt.subplot(3, 1, 1)
plt.title("Input Signal x[n]")
plt.stem(n, x)
plt.grid()

plt.subplot(3, 1, 2)
plt.title("Impulse response h[n]")
plt.stem(n, h)
plt.grid()

plt.subplot(3, 1, 3)
plt.title("y[n] = x[n] * h[n]")
plt.stem(np.arange(len(y)), y)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
'''
Filter realization using 6-point averaging, 6-point differencing equations. 
'''

import numpy as np
import matplotlib.pyplot as plt

# Sample input signal (analog-like)
np.random.seed(0)
n = np.linspace(0, 1, 200)
x = np.sin(2 * np.pi * 5 * n) + 0.5 * np.random.randn(len(n))  # Sine wave with noise

# 6-point Averaging Filter
def avg_filter(x):
    y = np.zeros_like(x)
    for i in range(5, len(x)):
        y[i] = np.sum(x[i-5:i+1]) / 6
    return y

# 6-point Differencing Filter
def diff_filter(x):
    y = np.zeros_like(x)
    for i in range(5, len(x)):
       y[i] = (x[i] - x[i-1] + x[i-2] - x[i-3] + x[i-4] - x[i-5]) / 6
    return y

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(n, x, label='Ordinal Signal', color='blue')
plt.title('Input Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(n, avg_filter(x), label='6-point Average Filter', color='blue')
plt.title('Input Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(n, diff_filter(x), label='# 6-point Differencing Filter', color='orange')
plt.title('Output Signal (Filtered)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```

```python
#   Consider a difference equation:
#       y[n] = (1 - a)y[n - 1] + a x[n]

#   Here, x[n] = [1, 2, 2, 10, 2, 2, 1]
#       a --> 0 < a < 1
#   Observation of y[n], effect of a.

import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 2, 10, 2, 2, 1])
n = np.arange(1, len(x) + 1)
a_values = [0.2, 0.5, 0.9]  # Different values of a

def difference_equation(x, a):
    y = [x[0]]
    for n in range(1, len(x)):
        cal = (1 - a) * y[n - 1] + a * x[n]
        y.append(cal)
    return np.array(y)

plt.figure(figsize=(10, 6))
plt.title('Difference Equation Output for Different a Values')
plt.plot(x, label='x[n]', color='gray')
plt.plot(difference_equation(x, a_values[0]), label=f'a = {a_values[0]}', color='red', marker='x')
plt.plot(difference_equation(x, a_values[1]), label=f'a = {a_values[0]}', color='blue', marker='x')
plt.xlabel('n (Index) x[n]')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show() 
```
