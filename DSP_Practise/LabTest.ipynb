Consider a difference equation:
    y[n] = (1 - a)y[n - 1] + a x[n]

Here, x[n] = [1, 2, 2, 10, 2, 2, 1]
    a --> 0 < a < 1
Observation of y[n], effect of a.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
x = np.array([1, 2, 2, 10, 2, 2, 1])
#alpha = np.linspace(0, 1, 5)
alpha = np.array([0.2])
for a in alpha:
    y = [x[0]]
    for n in range(1, len(x)):
        cal = (1 - a) * y[n - 1] + a * x[n]
        y.append(cal)
    
    plt.plot(y, label=f'alpha={a:.2f} y[n]', marker='o')
plt.plot(x, label='x[n]', color='gray')
plt.title('Effect of Alpha')
plt.grid(True)
plt.xlabel('n and X[i]')
plt.ylabel('X[n] and Y[i]')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter for continuous signal
f = 10 # frequency 10 Hz
n = np.linspace(0 , .2 , 150) # time vector for continuous time signal
x = np.sin(2 * np.pi * f * n) - np.cos(np.pi / 3) # continuous signal

# Plotting
plt.figure(figsize=(16 , 6))

# Demostrating for High Sampling frequency
plt.subplot(1,2,1)
plt.plot(n, x, label="np.sin(2 * np.pi * f * n) - np.cos(np.pi / 3)")
plt.title("Signal with noice")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
#alpha = np.linspace(0.2, 1, 3)

alpha = np.array([0.15])
plt.subplot(1, 2, 2)
for a in alpha:
    y = [x[0]]
    for n in range(1, len(x)):
        cal = (1 - a) * y[n - 1] + a * x[n]
        y.append(cal)
    
    plt.plot(y, label=f'alpha={a:.2f} y[n]', marker='o')
plt.plot(x, label='x[n]', color='gray')
plt.title('Effect of Alpha')
plt.grid(True)
plt.xlabel('n and X[i]')
plt.ylabel('X[n] and Y[i]')
plt.legend()

plt.show()
```
