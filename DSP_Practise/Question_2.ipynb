# Sampling and Aliasing in Digital Signal Processing

## üìå What is Sampling?

**Sampling** is the process of converting a continuous-time signal into a discrete-time signal by taking measurements (samples) at regular time intervals.

- If a signal is \( x(t) \), then the sampled signal is \( x[n] = x(nT_s) \), where \( T_s \) is the sampling period.
- The **sampling frequency** is \( f_s = \frac{1}{T_s} \).

### ‚úÖ Nyquist Theorem
To **avoid information loss**, the sampling frequency must be at least **twice the maximum frequency** of the signal:
\[
f_s \geq 2f_{\text{max}}
\]

---

## ‚ö†Ô∏è What is Aliasing?

**Aliasing** occurs when a signal is sampled at a rate lower than the Nyquist rate, causing different signals to become indistinguishable after sampling.

- High-frequency components get misinterpreted as lower frequencies.
- It leads to **distortion** in the reconstructed signal.

---

## üîç Python Demonstration of Sampling and Aliasing


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter for continuous signal
f = 10 # frequency 10 Hz
t_cont = np.linspace(0 , 1 , 1000) # time vector for continuous time signal
y_cont = np.sin(2 * np.pi * f * t_cont) # continuous signal

# Sampling rates
fs_high = 50 # Sampling rate higher than Nyquist
fs_low = 15 # Sampling rate lower than Nyquist

# Sampling time vectors
t_high = np.arange(0 , 1 , 1/fs_high) # time vector for high sampling rate
t_low = np.arange(0 , 1 , 1/fs_low) # time vector for low sampling rate
y_high = np.sin(2 * np.pi * f * t_high) # sampled signal at high rate
y_low = np.sin(2 * np.pi * f * t_low) # sampled signal at low rate

# Plotting
plt.figure(figsize=(16 , 6))

# Demostrating for High Sampling frequency
plt.subplot(1,2,1)
plt.plot(t_cont, y_cont, 'k--', label="Original Signal")
plt.stem(t_high, y_high, label='Sampled (fs=50Hz)')
plt.title("No Alising as fs >= 2.fmax")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

#Demonstrating Alaising Problem
plt.subplot(1,2,2)
plt.plot(t_cont, y_cont, 'k--', label="Original Signal")
plt.stem(t_low, y_low, label='Sampled (fs=15Hz)')
plt.title("Alising Problem")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

```
