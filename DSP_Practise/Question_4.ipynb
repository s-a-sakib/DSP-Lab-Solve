### Q4. Consider the analog signal 
###                 Xa(t) = 3 cos(100.pi.t)

###                 Sample it at 200 Hz and 75 Hz. Show the discrete signal  after 
###                 sampling. --> Realization

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal
def Xa(t):
    return 3 * np.cos(100 * np.pi * t)

# Sampling frequencies
fs1 = 200  # Sampling at 200 Hz
fs2 = 75   # Sampling at 75 Hz

# Time ranges for both sampling rates
t1 = np.arange(0, 0.2, 1/fs1)  # Time vector for fs = 200 Hz
t2 = np.arange(0, 0.2, 1/fs2)  # Time vector for fs = 75 Hz

# Sampled signals
x1 = Xa(t1)
x2 = Xa(t2)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.suptitle("Sampling of Xa(t) = 3cos(100πt) at Different Frequencies", fontsize=16)
# Plot for the continuous signal
t_continuous = np.linspace(0, 0.2, 1000)
x_continuous = Xa(t_continuous)
plt.plot(t_continuous, x_continuous, 'k-', label='Xa(t)', linewidth=2)
plt.fill_between(t_continuous, x_continuous, color='lightgray', alpha=0.5, label='Xa(t) Area')
plt.legend(loc='upper right')
plt.title("Continuous Signal Xa(t) = 3cos(100πt)", fontsize=14)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot the sampled signals
# Plot for 200 Hz sampling
plt.subplot(3, 1, 2)
plt.stem(t1, x1, basefmt=" ", linefmt='b-')
plt.title("Sampling of Xa(t) = 3cos(100πt) at 200 Hz")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

# Plot for 75 Hz sampling
plt.subplot(3, 1, 3)
plt.stem(t2, x2, basefmt=" ", linefmt='r-')
plt.title("Sampling of Xa(t) = 3cos(100πt) at 75 Hz")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
