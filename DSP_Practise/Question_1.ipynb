## Q1. Generate elementary signals like Unit Step, Ramp, Exponential, Sine and Cosine sequence.

### *Unit Step Signal*: Mathematically it is defined as:
                    f(x) = {
                        1 ; n >= 0
                        0 ; n < 0
                    }
### *Unit Ramp Signal*: Mathematically it is defined as:
                    f(x) = {
                        n ; n >= 0
                        0 ; n < 0
                    }
### *Exponential Signal*: Mathematically it is defined as:
                    f(x) = {
                        pow(a , n) ; n >= 0
                        0 ; n < 0
                    }
### *Sine Signal*: A sine wave is defined by the function:

###         y(t) = A⋅sin(2πft + ϕ)
###         Where:

###         𝐴 = Amplitude (peak value)

###         𝑓 = Frequency (in Hz, cycles per second)

###         𝑡 = Time (in seconds)

###        𝜙 = Phase shift (in radians)

### *Cosine Signal*: A cosine wave is defined by the function:

###         y(t) = A⋅cos(2πft + ϕ)
###         Where:

###         𝐴 = Amplitude (peak value)

###         𝑓 = Frequency (in Hz, cycles per second)

###         𝑡 = Time (in seconds)

###        𝜙 = Phase shift (in radians)

```python
import numpy as np
import matplotlib.pyplot as plt

def unit_step(n):
    return np.where(n >= 0, 1, 0)

def ramp(n):
    return np.where(n >= 0, n, 0)

def exponential(n, a=0.5):
    return np.exp(a * n)

def sine_wave(n, freq=1, fs=10):
    return np.sin(2 * np.pi * freq * n / fs)

def cosine_wave(n, freq=1, fs=10):
    return np.cos(2 * np.pi * freq * n / fs)


# Discrete time index
n = np.arange(-10, 10, 1)

# Generate signals
signals = {
    "Unit Step": unit_step(n),
    "Ramp": ramp(n),
    "Exponential": exponential(n),
    "Sine": sine_wave(n),
    "Cosine": cosine_wave(n)
}

# Plot signals
plt.figure(figsize=(12, 8))
for i, (title, signal) in enumerate(signals.items(), 1):
    plt.subplot(3, 2, i)
    plt.stem(n, signal, basefmt="b-")
    plt.title(title)
    plt.xlabel("n")
    plt.ylabel("Amplitude")
    plt.grid()

plt.tight_layout()
plt.show()
```
