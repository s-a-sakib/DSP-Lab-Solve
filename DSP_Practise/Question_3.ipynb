A discrete-time sinusoid is defined as:

    n: sample index
    Ï•: phase offset

In discrete-time signals, frequencies are periodic with period 
2Ï€. The number of zero crossings per unit time increases with

This alternates between +1 and -1 â€” the fastest possible rate of change in a discrete signal (every sample changes sign).

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 10, 1)
omega1 = np.pi
omega2 = 0.5 * np.pi

wave1 = np.cos(n * omega1)
wave2 = np.cos(n * omega2)

wave3 = wave1 + wave2

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.stem(n, wave1, 'r-', label='omega = pi')
plt.title('Cosine Waves with Different Frequencies')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.stem(n, wave2, 'g-', label='omega = 0.5 * pi')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
