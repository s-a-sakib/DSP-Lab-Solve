### Q5. Consider the analog signal:
### Xa(t) = 3cos(2000.pi.t) + 5sin(6000.pi.t)+10cos(12000.pi.t)

### Show the effect of sampling.

    If Sampling Rate:

    fs​>12000 Hz → No aliasing, signal preserved.
    fs​=12000 Hz → Barely no aliasing (risky).
    fs​<12000 Hz → Aliasing occurs (high frequencies fold into lower ones).
    Example:

    Sampling at 8000 Hz:

    1000 Hz → OK
    3000 Hz → OK
    6000 Hz → aliases to 2000 Hz.

```python
import numpy as np
import matplotlib.pyplot as plt

A = [3, 5, 10]  # Amplitude
f = [1000, 3000, 6000] # Frequency

t_cont = np.linspace(0, 0.005, 10000)

def X(A1, A2, A3, F1, F2, F3, t_cont):
    return A1 * np.cos(2 * np.pi * F1 * t_cont) + A2 * np.sin(2 * np.pi * F2 * t_cont) + A3 * np.cos(2 * np.pi * F3 * t_cont)

# Create the signal 
signal = X(A[0], A[1], A[2], f[0], f[1], f[2], t_cont)
# Plot the signal
plt.figure(figsize=(10, 5))
plt.subplot(4, 1, 1)
plt.title(r'Sampling of $X_a(t) = 3\cos(2000\pi t) + 5\sin(6000\pi t) + 10\cos(12000\pi t)$')
plt.plot(t_cont, signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()

# Creating sampled signal greater than the Nyquist rate for DTS
sampling_rate = 20000
t_cont = np.arange(0, 0.005, 1/sampling_rate)

signal = X(A[0], A[1], A[2], f[0], f[1], f[2], t_cont)
plt.subplot(4, 1, 2)
plt.title('Sampled Signal at 20000Hz')  # Fixed subtitle to suptitle
plt.stem(t_cont, signal)
plt.plot(t_cont, signal, 'r', alpha=0.5)  # Add a line plot for better visibility
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()

# Creating sampled signal less than the Nyquist rate for DTS
sampling_rate = 10000
t_cont = np.arange(0, 0.005, 1/sampling_rate)
signal = X(A[0], A[1], A[2], f[0], f[1], f[2], t_cont)
plt.subplot(4, 1, 3)
plt.title('Sampled Signal at 10000Hz')  # Fixed subtitle to suptitle
plt.stem(t_cont, signal)
plt.plot(t_cont, signal, 'r', alpha=0.5)  # Add a line plot for better visibility
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```
